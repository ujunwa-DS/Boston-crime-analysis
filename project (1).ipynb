Introduction/Problem Statement
Considering the risk on life and prosperities, there is a need to have a factual information about crimes.

Crime incident reports are provided by Boston Police Department (BPD) to document the initial details surrounding an incident to which BPD officers respond since from 2015 to 2018. This is a dataset containing records from the new crime incident report system, which includes a reduced set of fields focused on capturing the type of incident as well as when and where it occurred.

The dataset had 16 columns which had : 
1. Location 
2. Time 
3. Identifier - incident number 
4. Offense description
5. Incident number
6. Street
7. Month of occurence 

BACKGROUND 

Crime is an unlawful act punishable by a state or other authorties. There are two major categories of crime which is 
a. Violent crime: a crime which leads to loss of life. Example here include homicides, unlawful shooting, rape, aggravated assult etc. 
b. Non violent crime: Crimes with no intent of causing loss of life. examples include, Verbal abuse, property damage, sick assist(use of alcohol by minors) etc. From 1980 to 2020, there have been a tremendous decrease on how often violent crime occurs (wikipedia) while non-violent crime has not shown any positive trend. 
All these crimes are categorized into 6 types of crime depending on where it occurred, places it affects and what it affects. 
1. Crime to life
2. Crime against properties
3. Crime against morality
4. Organized crime
5. Hate crimes
6. White collar crimes

The Boston police has a total number of 68075 report from June 14, 2015 and continue to September 3, 2018. It shows that a total of about 59 reports are made on daily basis in all the Street of Boston
